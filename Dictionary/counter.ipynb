```python
def vcnt(s):
    v = ['a','e','i','o','u']
    cnt=0
    for i in s:
        if i in v:
            cnt+=1
    return s,cnt
```

```python
vcnt('mississippi')
```

```python
vcnt('welcome')
```

```python
bag_of_words = ['welcome', 'to', 'the', 'world', 'of', 'python', 'welcome', 'to', 'the', 'world', 'of', 'python']
for i in map(vcnt, bag_of_words):
    print(i)
```

```python
for w,c in map(vcnt, bag_of_words):
    print(w+" "+str(c))
```

```python
bag_of_words = ['welcome', 'to', 'fullstackpython', 'welcome', 'to', 'dsa']
```

```python
def vcnt2(s):
    v = ['a','e','i','o','u']
    cnt=0
    for i in s:
        if i in v:
            cnt+=1
    return cnt>=3
```

```python
for i in filter(vcnt2, bag_of_words):
    print(i)
```

```python
for i in map(vcnt,filter(vcnt2, bag_of_words)):
    print(i)
```
