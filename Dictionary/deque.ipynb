```python
# deque is a double-ended queue that allows you to add and remove elements from both ends efficiently. It is part of the collections module in Python.
# The deque class provides methods for adding and removing elements from both ends of the queue, as well as other useful methods for working with deques.
# Here are some of the key methods of the deque class:
# - append(x): Add x to the right end of the deque.
# - appendleft(x): Add x to the left end of the deque.
# - pop(): Remove and return an element from the right end of the deque. If the
#   deque is empty, raises an IndexError.
# - popleft(): Remove and return an element from the left end of the deque. If
#   the deque is empty, raises an IndexError.

import collections
```
